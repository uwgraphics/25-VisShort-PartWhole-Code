```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set seed for reproducibility
rng = np.random.default_rng(123456789)

parts = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
charts = ["stackbar"]
alignments = ["aligned", "non-aligned"]
selectedParts = range(11, 51, 1)
sets = ['set1', 'set2', 'set3', 'set4']
sections = [1]
numberParticipants = 60
questionsPerParticipant = len(selectedParts) * len(charts) * 2
totalQuestions = numberParticipants * questionsPerParticipant

columns = ['Set', 'SelectedPart', 'SelectedLabel', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'A_value', 'B_value', 'C_value', 'D_value', 'E_value', 'F_value', 'G_value', 'A_index', 'B_index', 'C_index', 'D_index', 'E_index', 'F_index', 'G_index']
df = pd.DataFrame(columns=columns)

remainingPartsTotal = []
remainingPartsSets = []
for set in sets:
  for selectedPart in selectedParts:
    maximum = rng.integers(101, 999)
    remainingParts = []
    total = selectedPart
    order = rng.permutation(range(7)).astype(int)

    for j in range(5):
      upperBound = min((7-j)*5, (100 - total))
      remainingPart = np.round(rng.beta(a=2, b=10) * (100 - total)).astype(int)
      remainingParts.append(remainingPart)
      total = selectedPart + sum(remainingParts)

    remainingParts.append(100 - total)

    rng.shuffle(remainingParts)
    remainingPartsTotal.extend(remainingParts)

    selected = rng.integers(0, 7)
    remainingParts.insert(selected, selectedPart)
    selectedLabel = parts[selected]

    remainingPartsSets.append(remainingParts)
    selectedPartIndex = parts.index(selectedLabel)

    order = rng.permutation(range(7)).astype(int)
    while order[selectedPartIndex] in [0, 6] or np.sum(np.array(remainingParts)[order[:np.where(np.isin(order, selectedPartIndex))[0][0]]]) in [0, 25, 50, 75, 100]:
      order = rng.permutation(range(7)).astype(int)

    rowDict = {
        'Set': set,
        'SelectedPart': selectedPart,
        'SelectedLabel': selectedLabel,
        'A': remainingParts[0],
        'B': remainingParts[1],
        'C': remainingParts[2],
        'D': remainingParts[3],
        'E': remainingParts[4],
        'F': remainingParts[5],
        'G': remainingParts[6],
        'A_value': np.round(remainingParts[0]/100 * maximum).astype(int),
        'B_value': np.round(remainingParts[1]/100 * maximum).astype(int),
        'C_value': np.round(remainingParts[2]/100 * maximum).astype(int),
        'D_value': np.round(remainingParts[3]/100 * maximum).astype(int),
        'E_value': np.round(remainingParts[4]/100 * maximum).astype(int),
        'F_value': np.round(remainingParts[5]/100 * maximum).astype(int),
        'G_value': np.round(remainingParts[6]/100 * maximum).astype(int),
        'A_index': order[0],
        'B_index': order[1],
        'C_index': order[2],
        'D_index': order[3],
        'E_index': order[4],
        'F_index': order[5],
        'G_index': order[6]
    }
    df = pd.concat([df, pd.DataFrame([rowDict])], ignore_index=True)



def shuffle_questions(group):
    group['QuestionID'] = rng.permutation(group['QuestionID'])
    return group

qid = 0
columns = ['QuestionID', 'UserID', 'Set', 'SelectedPart', 'SelectedLabel', 'ChartType', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'A_value', 'B_value', 'C_value', 'D_value', 'E_value', 'F_value', 'G_value', 'A_index', 'B_index', 'C_index', 'D_index', 'E_index', 'F_index', 'G_index']
questions = pd.DataFrame(columns=columns)
for user in range(numberParticipants):

# engagements
#   rowDict = {
#       'QuestionID': qid,
#       'UserID': user,
#       'Set': "engagement",
#       'SelectedPart': 50,
#       'SelectedLabel': 'D',
#       'ChartType': "pie",
#       'A': 50, 'B': 0, 'C': 0, 'D': 50, 'E': 0, 'F': 0, 'G': 0,
#       'A_value': 50, 'B_value': 0, 'C_value': 0, 'D_value': 50, 'E_value': 0, 'F_value': 0, 'G_value': 0,
#       'A_index': 0, 'B_index': 1, 'C_index': 2, 'D_index': 3, 'E_index': 4, 'F_index': 5, 'G_index': 6
#   }
#   questions = pd.concat([questions, pd.DataFrame([rowDict])], ignore_index=True)
#   qid += 1

#   rowDict = {
#       'QuestionID': qid,
#       'UserID': user,
#       'Set': "engagement",
#       'SelectedPart': 50,
#       'SelectedLabel': 'C',
#       'ChartType': "stackbar",
#       'A': 0, 'B': 0, 'C': 50, 'D': 0, 'E': 0, 'F': 0, 'G': 50,
#       'A_value': 0, 'B_value': 0, 'C_value': 50, 'D_value': 0, 'E_value': 0, 'F_value': 0, 'G_value': 50,
#       'A_index': 0, 'B_index': 1, 'C_index': 2, 'D_index': 3, 'E_index': 4, 'F_index': 5, 'G_index': 6
#   }
#   questions = pd.concat([questions, pd.DataFrame([rowDict])], ignore_index=True)
#   qid += 1

#   rowDict = {
#       'QuestionID': qid,
#       'UserID': user,
#       'Set': "engagement",
#       'SelectedPart': 100,
#       'SelectedLabel': 'G',
#       'ChartType': "bar",
#       'A': 0, 'B': 0, 'C': 0, 'D': 0, 'E': 0, 'F': 0, 'G': 100,
#       'A_value': 0, 'B_value': 0, 'C_value': 0, 'D_value': 0, 'E_value': 0, 'F_value': 0, 'G_value': 100,
#       'A_index': 2, 'B_index': 6, 'C_index': 0, 'D_index': 4, 'E_index': 5, 'F_index': 3, 'G_index': 1
#   }
#   questions = pd.concat([questions, pd.DataFrame([rowDict])], ignore_index=True)
#   qid += 1
#   rowDict = {
#       'QuestionID': qid,
#       'UserID': user,
#       'Set': "engagement",
#       'SelectedPart': 100,
#       'SelectedLabel': 'A',
#       'ChartType': "table",
#       'A': 100, 'B': 0, 'C': 0, 'D': 0, 'E': 0, 'F': 0, 'G': 0,
#       'A_value': 100, 'B_value': 0, 'C_value': 0, 'D_value': 0, 'E_value': 0, 'F_value': 0, 'G_value': 0,
#       'A_index': 0, 'B_index': 4, 'C_index': 6, 'D_index': 3, 'E_index': 1, 'F_index': 2, 'G_index': 5
#   }
#   questions = pd.concat([questions, pd.DataFrame([rowDict])], ignore_index=True)
#   qid += 1

  for section in sections:
    for chart in charts:
      for alignment in alignments:
        for selectedPart in selectedParts:
          setNumber = rng.choice(sets[:4])
          row = df[(df['Set'] == setNumber) & (df['SelectedPart'] == selectedPart)]

          rowDict = {
              'QuestionID': qid,
              'UserID': user,
              'Set': setNumber,
              'SelectedPart': int(row["SelectedPart"]),
              'SelectedLabel': row["SelectedLabel"].tolist()[0],
              'ChartType': chart,
              'Alignment': alignment,
              'A': int(row['A']),
              'B': int(row['B']),
              'C': int(row['C']),
              'D': int(row['D']),
              'E': int(row['E']),
              'F': int(row['F']),
              'G': int(row['G']),
              'A_value': int(row['A_value']),
              'B_value': int(row['B_value']),
              'C_value': int(row['C_value']),
              'D_value': int(row['D_value']),
              'E_value': int(row['E_value']),
              'F_value': int(row['F_value']),
              'G_value': int(row['G_value']),
              'A_index': int(row['A_index']),
              'B_index': int(row['B_index']),
              'C_index': int(row['C_index']),
              'D_index': int(row['D_index']),
              'E_index': int(row['E_index']),
              'F_index': int(row['F_index']),
              'G_index': int(row['G_index'])
          }
          questions = pd.concat([questions, pd.DataFrame([rowDict])], ignore_index=True)
          qid += 1

# Function to modify indices based on SelectedLabel
def modify_indices(row):
    if row['Alignment'] == 'aligned':
        label = row['SelectedLabel']
        label_index = label + '_index'
        label_index_value = row[label_index]

        # Modify the corresponding index value to 0
        row[label_index] = 0

        # Shift other indices accordingly
        for col in ['A_index', 'B_index', 'C_index', 'D_index', 'E_index', 'F_index', 'G_index']: # Adjust column names if more indices are present
            if col != label_index:
                current_index = row[col]
                if current_index <= label_index_value:
                    row[col] += 1
    return row

# Apply the function row-wise to modify indices
questions = questions.apply(modify_indices, axis=1)

# Shuffle Questions
questions = questions.groupby('UserID', group_keys=False).apply(shuffle_questions).sort_values(by='QuestionID')


selected_columns = ['QuestionID', 'UserID', 'SelectedPart', 'SelectedLabel', 'ChartType', 'A_index', 'B_index', 'C_index', 'D_index', 'E_index', 'F_index', 'G_index', 'Alignment']
selected_df = questions[selected_columns]
selected_df
# sns.histplot(remainingPartsTotal, bins=40)
# plt.title('Distribution of Other Parts')
# plt.xlabel('Values (Percent of Whole)')
# plt.ylabel('Frequency')
# plt.show()
```
